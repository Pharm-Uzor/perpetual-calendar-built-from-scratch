```python
# create data structures storing names of days, months and number of days per month

days = list('MTWTFSS')

days_per_month = {1:31, 2:28, 3:31, 4:30, 5:31, 6:30, 7:31, 8:31, 9:30, 10:31, 11:30, 12:31}

month_name = {1:'Jan', 2:'Feb', 3:'Mar', 4:'Apr', 5:'May', 6:'Jun', 7:'Jul', 8:'Aug', 9:'Sep', 10:'Oct', 11:'Nov', 12:'Dec'}

#set the number of days in february to 29 if year is leap year
if (int(year) + 1) % 4 == 0:
    days_per_month[2] = 29 
```

```python
# define utility functions 
# 1) implementation of Zeller's Congruence algorithm for determining the first week day of any given year
# 2) print empty calendar layout
# 3) fill up months with days 

def First_day(year): 
    D_ = int(year[2:]) # select last two digits of year
    C_ = int(year[:2]) # select first two digits of year
    F = 29 + D_ + (D_//4) + (C_//4) - 2 * C_
    if F % 7 == 0:
        x_ = 6
    else:
        x_ = F % 7 - 1
    return x_
       

def print_layout(mth, first_day):
    if first_day != 0:  # print a new line if first day of month is not Monday to create a uniform layout althrough
        print('\n')
    print(f'           {month_name[mth]}', end = '') # print header month names for each month of year
    print()
    for D in days:   # headers for names of days of week
        print(D + '  ', end = ' ')
    print('\n')
    print('  ' * first_day * 2, end = '') # skip to the first day of the month to commence filling with days
    
def fill_month(ndays_per_mth, first_day):
    day = 0
    while (day) < ndays_per_mth:
        if day < 9:   # to align the position of single digits and double digits, add a white space to single digits
            print(f'{day + 1}  ', end = ' ') # print single digit days
        else:
            print(f'{day + 1}', end = '  ') # print double digit days
        day += 1  # update the day of the month
            
        first_day += 1  # update the day of the week 
        if first_day % 7 == 0:
            first_day = 0 # reset day of week to zero at the beginning of a new week
            print('\n')
    print('\n')
    return first_day # return info on next week day
    
```

```python
print(f'    Calender for {str(int(year) + 1)}')
print('\n')
print('Written by Uzodimma Igodo E.', end = '')
print('\n')

day_of_week = First_day(year) # retreive the first week day of the year

for mth, ndays_per_mth in days_per_month.items():  # iteratively build months for the given year
    print_layout(mth, day_of_week)  # 1) build an empycalender structure/layout for a specific month
    day_of_week = fill_month(ndays_per_mth, day_of_week)   # 2) fill the empty month with days
```
